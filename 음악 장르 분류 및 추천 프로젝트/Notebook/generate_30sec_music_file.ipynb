```python
from pydub import AudioSegment
import random
import os
```

```python
genre_dict = {'bollypop': 'bp', 'carnatic': 'ca', 'ghazal': 'gz', 'semiclassical': 'scl', 'sufi': 'su'}
```

```python
# 랜덤 30초 자르기
def random_music_30cutter(music):
    if music.duration_seconds <= 30:
        return music
    start = random.randint(0, music.duration_seconds - 30)
    return music[start * 1000 : start * 1000 + 30000]
```

```python
# 중간 30초 자르기
def middle_music_30cutter(misic):
    if music.duration_seconds <= 30:
        return music
    start = (music.duration_seconds - 30) // 2
    return music[start * 1000 : start * 1000 + 30000]
```

```python
for genre, genre_short in genre_dict.items():
    # 폴더 생성
    os.mkdir(f'../data/genrenew30/{genre}')
    
    for n in range(100):
        file = genre_short + str(n + 1).zfill(2)
        music = AudioSegment.from_mp3(f'../data/genrenew/{genre}/{file}.mp3')

        # 원하는 함수 적용
        result = middle_music_30cutter(music)
        
        result.export(f'../data/genrenew30/{genre}/{file}.mp3', format='mp3')
        print(f'../data/genrenew/{genre}/{file}.mp3 finish')
```
